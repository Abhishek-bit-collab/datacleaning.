```python
import pandas as pd

# Load dataset (replace filename with your actual file name)
df = pd.read_csv("diabetes.csv")

# Now check columns
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap of correlations
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()

# Boxplot for BMI
sns.boxplot(x=df['BMI'])
plt.show()

# Scatter plot: Age vs Glucose
sns.scatterplot(x='Age', y='Glucose', data=df)
plt.show()

# Histogram: Distribution of BloodPressure
sns.histplot(df['BloodPressure'], bins=30, kde=True)
plt.show()
```

```python
# Outcome count plot
sns.countplot(x='Outcome', data=df)
plt.show()

# Histogram of Glucose
sns.histplot(df['Glucose'], bins=30, kde=True)
plt.show()

# Boxplot of Insulin
sns.boxplot(x=df['Insulin'])
plt.show()

# Scatter plot Age vs BMI
sns.scatterplot(x='Age', y='BMI', data=df, hue='Outcome')
plt.show()
```

## Insights from Diabetes Dataset

- Glucose distribution shows many patients with high values, strongly linked to diabetes outcome.  
- Insulin column has extreme outliers, which may affect modeling.  
- BMI is positively related to diabetes risk, especially in older patients.  
- Outcome count plot shows imbalance: more non-diabetic cases than diabetic.
